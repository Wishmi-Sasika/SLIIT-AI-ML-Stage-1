# Import Data

```python
import pandas as pd
url = "https://raw.githubusercontent.com/selva86/datasets/master/Ionosphere.csv"
df = pd.read_csv (url)
df.head()
```

# Identifying the dependant and independant variables

```python
x = df.drop(['Class'], axis = 1)
y = df['Class']
```

```python
# Dependant variable
print (y)
```

```python
# Independant variable
print (x)
```

# Splitting the dataset into Training and Testing datasets

draw.drawio.png

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (x, y, test_size = 0.2)
```

# Available packages in Colab

```python
pip list
```

# Importing dependencies

```python
from keras.models import Sequential, load_model
from keras.layers import Dense, Dropout
```

# Building the model

image.png

```python
model = Sequential()

# First hidden layer
# Normally, for hidden layers, we use the relu activation function.
# In relu activation function, the minaze values will be assigned into 0.
model.add (Dense (units = 10, activation = 'relu', input_dim = len(x_train.columns)))

# Second hidden layer
model.add (Dense (units = 5, activation = 'relu'))

# Output layer
# In sigmoid activation function, we get a probabilistic value as the output.
model.add (Dense (units = 1, activation = 'sigmoid'))
```

```python
model.summary()
```

# Compiling the model

```python
# Based on the loss, we can learn that whether the model is training poperly or not.
# In here, it is binary_crossentropy because, this is a binary classification.
model.compile (loss = 'binary_crossentropy', optimizer = 'sgd', metrics = 'accuracy')
```

```python
model.fit (x_train, y_train, validation_data = (x_test, y_test), epochs = 100, batch_size = 32)
```

```python
model.save ('weights.h5')
```
