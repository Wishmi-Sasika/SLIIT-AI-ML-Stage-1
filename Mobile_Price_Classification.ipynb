# Importing libraries

```python
import pandas as pd
import nltk
from sklearn.model_selection import train_test_split
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
```

# Importing the dataset

```python
df = pd.read_csv ("/content/drive/MyDrive/Colab Notebooks/Data/Mobile_Price_Classification.csv", encoding = "latin-1")
df.head()
```

```python
df.shape
```

```python
df.groupby ('price_range').describe()
```

# Removing duplicates

```python
df = df.drop_duplicates (keep = 'first')
df.groupby ('price_range').describe()
```

# Splitting the dataset

```python
x = df.drop('price_range', axis=1)
y = df['price_range']
```

```python
x_train, x_test, y_train, y_test = train_test_split (x, y, test_size = 0.25, random_state = 2)
```

# Building the model

```python
model = Sequential()

model.add (Dense (units = 8, activation = 'relu', input_dim = len(x_train.columns)))
model.add (Dense (units = 4, activation = 'relu'))
model.add (Dense (units = 1, activation = 'sigmoid'))
```

```python
model.summary()
```

# Compiling the model

```python
model.compile (loss = 'binary_crossentropy', optimizer = 'sgd', metrics = 'accuracy')
```

# Fitting the model

```python
model.fit (x_train, y_train, validation_data = (x_test, y_test), epochs = 100, batch_size = 32)
```

# Saving the weights file

```python
model.save ('weights_Task1.h5')
```
